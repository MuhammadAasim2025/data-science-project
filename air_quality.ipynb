```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the 'Data Value' column
df['Data Value Normalized'] = scaler.fit_transform(df[['Data Value']])

display(df[['Data Value', 'Data Value Normalized']].head())
```

```python
df['Day_of_Week'] = df['Start_Date'].dt.dayofweek
df['Day_of_Year'] = df['Start_Date'].dt.dayofyear

display(df[['Start_Date', 'Day_of_Week', 'Day_of_Year']].head())
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, accuracy_score

```

```python
# Convert text columns to numbers
le = LabelEncoder()
cols_to_encode = ['Name', 'Geo Place Name', 'Measure']
for col in cols_to_encode:
    df[col + '_Encoded'] = le.fit_transform(df[col].astype(str))

# Define Features (X) - What we use to predict
X = df[['Name_Encoded', 'Geo Place Name_Encoded', 'Year', 'Month']]

# ---------------------------------------------------------
# SCENARIO 1: REGRESSION (Predicting the exact Data Value)
# ---------------------------------------------------------
y_reg = df['Data Value'] # Target is a number

# Split into Training (80%) and Testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# Create and Train the Regressor
reg_model = DecisionTreeRegressor()
reg_model.fit(X_train, y_train)

# Test the model
predictions = reg_model.predict(X_test)
print(f"Regression Model MSE: {mean_squared_error(y_test, predictions):.2f}")


# ---------------------------------------------------------
# SCENARIO 2: CLASSIFICATION (Predicting High vs Low)
# ---------------------------------------------------------
# Create a category: 1 if Value is above average, 0 if below
threshold = df['Data Value'].mean()
y_class = (df['Data Value'] > threshold).astype(int)

# Split into Training and Testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)

# Create and Train the Classifier
clf_model = DecisionTreeClassifier()
clf_model.fit(X_train, y_train)

# Test the model
accuracy = accuracy_score(y_test, clf_model.predict(X_test))
print(f"Classification Model Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 1. Prepare data for Regression (Predicting the exact Data Value)
X = df[['Name_Encoded', 'Geo Place Name_Encoded', 'Year', 'Month']]
y_reg = df['Data Value']

# 2. Split and Train Random Forest
X_train, X_test, y_train, y_test = train_test_split(X, y_reg, test_size=0.2, random_state=42)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 3. Get the numbers for Section 6 (Results)
y_pred = rf_model.predict(X_test)
print(f"Regression R-Squared Accuracy: {r2_score(y_test, y_pred):.4f}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred):.4f}")
```
