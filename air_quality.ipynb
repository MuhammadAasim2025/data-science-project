```python
 df=df.drop(columns=[
    "Message",
    "Geo Join ID",
    "Measure Info"
], errors="ignore")
```

```python
df.dropna(subset=["Data Value"])
```

```python
df = df.dropna(subset=[
    "Name",
    "Geo Place Name",
    "Time Period",
    "Start_Date"
])
```

```python
df
```

```python
df["Start_Date"] = pd.to_datetime(df["Start_Date"], errors="coerce")
df = df.dropna(subset=["Start_Date"])
```

```python
df["Year"] = df["Start_Date"].dt.year
df["Month"] = df["Start_Date"].dt.month
```

```python
df
```

```python
text_cols = ["Name", "Geo Place Name", "Geo Type Name", "Time Period"]

for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df
```

```python
display(df.iloc[0:31])
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=["Name", "Geo Place Name", "Geo Type Name", "Time Period"],
    drop_first=True
)
```

```python
df
```

```python
df_encoded = df_encoded.drop(columns=["Start_Date"], errors="ignore")
```

```python
df_encoded
```

```python
df["Data Value"] = pd.to_numeric(df["Data Value"], errors="coerce")
df = df.dropna(subset=["Data Value"])
```

```python
df["Data Value"] = (
    pd.to_numeric(df["Data Value"], errors="coerce")
    .round(2)
)
df = df.dropna(subset=["Data Value"])
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the 'Data Value' column
df['Data Value Normalized'] = scaler.fit_transform(df[['Data Value']])

display(df[['Data Value', 'Data Value Normalized']].head())
```

```python
df['Day_of_Week'] = df['Start_Date'].dt.dayofweek
df['Day_of_Year'] = df['Start_Date'].dt.dayofyear

display(df[['Start_Date', 'Day_of_Week', 'Day_of_Year']].head())
```
